```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
# Brooklyn: Define a function to get name and height information and create as a dataframe
def BKInfo(URL):
    page = requests.get(URL, verify=False)
    soup = BeautifulSoup(page.content, 'html.parser')
    # web scraping for names
    NameList = []
    NList = []
    Name = soup.find_all("h3", class_=False, id=False)
    for name in Name:
      NList.append(name.get_text().strip('\n'))
    # removing coaching staff tag
    NameList = NList[:(len(NList)-2)]
    # web scraping for heights  
    HeightList = []
    InchList = []
    Height = soup.find_all('span', class_= "sidearm-roster-player-height")
    for height in Height:
      HeightList.append(height.get_text())
    # convert feet and inch to inch only
    for x in HeightList:
      a = x.replace('"','')
      height = a.split("'")
      feet = int(height[0])
      inch = int(height[1])
      TotInch = inch + 12*feet
      InchList.append(TotInch)
# Since the HeightList will returns twice of the whole real list, only needs the first part
    m = int(len(InchList)/2)
    InchList = InchList[0:m]
# create a dictionary to combine these two lists and create a data frame
    a = {'Name': NameList, 'Height': InchList}
    df = pd.DataFrame(a)
    return df
```

```python
# Baruch: Define a function to get name and height info and create as a dataframe
def BInfo(URL):
    page = requests.get(URL, verify=False)
    soup = BeautifulSoup(page.content, 'html.parser')
    # web scraping for names
    NameList = []
    NList = []
    Name = soup.find_all("h3", class_=False, id=False)
    for name in Name:
      NList.append(name.get_text().strip('\n'))
      # removing coaching staff tag
      NameList = NList[:(len(NList)-2)]
    # web scraping for heights  
    HeightList = []
    InchList = []
    Height = soup.find_all('span', class_= "sidearm-roster-player-height")
    for height in Height:
      HeightList.append(height.get_text())
    # convert feet and inch to inch only
    for x in HeightList:
      a = x.replace('"','')
      height = a.split("'")
      feet = int(height[0])
      inch = int(height[1])
      TotInch = inch + 12*feet
      InchList.append(TotInch)
# Since the HeightList will returns twice of the whole real list, only needs the first part
    m = int(len(InchList)/2)
    InchList = InchList[0:m]
# create a dictionary to combine these two lists and create a data frame
    a = {'Name': NameList, 'Height': InchList}
    df = pd.DataFrame(a)
    return df
```

```python
# York & Queens: Define a function to get name and height info and create as a dataframe
def YQInfo(URL):
    page = requests.get(URL, verify=False)
    soup = BeautifulSoup(page.content, 'html.parser')
    NameList = []
    Name = soup.find_all('td', class_= "sidearm-table-player-name")
    for name in Name:
      NameList.append(name.get_text().strip('\n'))
    HeightList = []
    Height = soup.find_all('td', class_= "height")
    InchList = []
    for height in Height:
        HeightList.append(height.get_text())
# convert feet and inch to inch only
    for x in HeightList:
      a = x.replace('-','')
      feet = int(a[0])
      if len(a) == 2:
        inch = int(a[1])
        TotInch = inch + 12*feet
      else:
        TotInch = 10 + 12*feet
      InchList.append(TotInch)
# create a dictionary to combine these two lists and create a data frame
    a = {'Name': NameList, 'Height': InchList}
    df = pd.DataFrame(a)
    return df
```

```python
#John Jay: Define a function to get name and height info and create as a dataframe
def JInfo(URL):
    page = requests.get(URL, verify=False)
    soup = BeautifulSoup(page.content, 'html.parser')
    FirstList = []
    LastList = []
    NameList = []
    FName = soup.find_all('span', class_= "sidearm-roster-player-first-name")
    LName = soup.find_all('span', class_= "sidearm-roster-player-last-name")
    HeightList = []
    Height = soup.find_all('span', class_= "sidearm-roster-player-height")
    InchList = []
    for height in Height:
        HeightList.append(height.get_text())
    for x in HeightList:
      a = x.replace('"','')
      height = a.split("'")
      feet = int(height[0])
      inch = int(height[1])
      TotInch = inch + 12*feet
      InchList.append(TotInch)

# Since the HeightList will returns twice of the whole real list, only needs the first part
    m = int(len(InchList)/2)
    InchList = InchList[0:m]
    for name in FName:
      FirstList.append(name.get_text())
    for name in LName:
      LastList.append(name.get_text())
# combine first name and last name to one list
    N = ''
    for r in range(0,m):
      N = FirstList[r] + ' ' + LastList[r]
      NameList.append(N)
# create a dictionary to combine these two lists and create a data frame
    a = {'Name': NameList, 'Height': InchList}
    df = pd.DataFrame(a)
    return df
```

```python
# 1. Men Volleyball

# get Brooklyn Men's Volleyball info
BKMenV = BKInfo('https://www.brooklyncollegeathletics.com/sports/mens-volleyball/roster/2019')
BKMenV

# get Baruch Men's Volleyball info
BMenV = BInfo('https://athletics.baruch.cuny.edu/sports/mens-volleyball/roster')
BMenV 

# get York Men's Volleyball info
YMenV = YQInfo('https://yorkathletics.com/sports/mens-volleyball/roster')
YMenV

# All names and heights data of Men's Volleyball, and save as new data frame
MV = pd.concat([BKMenV,BMenV,YMenV])

# Storing the data by heights permanent and in a descending order
# Name in ascending order and reset index
MV.sort_values(by=['Height', 'Name'], ascending= [False,True], inplace = True)
MV.reset_index(inplace=True)

# delete the original index and stor with name
MV.drop('index', inplace=True, axis=1)

MV
```

```python
# find average height
MVAvg = MV['Height'].mean()

print("The average height of Men's Volleyball players is {} inch".format(MVAvg))
```

```python
# find the 5 tallest players names and heights
TallMV = MV.head()
print("The Men's Volleyball 5 tallest are: ")
TallMV
```

```python
# find the 5 shortest plays names and heights
ShortMV = MV.tail()
print("The Men's Volleyball 5 shortest are: ")
ShortMV
```
